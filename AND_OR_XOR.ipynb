```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "a": [0,0,1,1],
    "b": [0,1,0,1],
    "a.b":[0,0,0,1]}
df = pd.DataFrame(data)
```

```python
df
```

```python
x = np.array(df.drop('a.b',axis=1))
y = np.array(df['a.b'])
```

```python
x
```

```python
class Layer:
    def __init__(self):
        self.input = None
        self.output = None
    def forward(self):
        pass
    def backward(self):
        pass
```

```python
class Dense(Layer):
    def __init__(self, input_size, output_size):
        self.weights = np.random.randn(output_size,input_size)
        self.bias = np.random.randn(output_size, 1)
    def forward(self, input):
        self.input = input
        return np.dot(self.weights, self.input)+ self.bias
    def backward(self, output_gradient, learning_rate):
        weights_gradient = np.dot(output_gradient, self.input.T)
        input_gradient = np.dot(self.weights.T, output_gradient)
        self.weights -= learning_rate* weights_gradient
        self.bias -= learning_rate*output_gradient
        return input_gradient
```

```python
def predict(network, input):
    output = input 
    for layer in network:
        output = layer.forward(output)
    return output
```

```python
def train(network, loss, loss_gradient, x_train, y_train, epochs = 1000, learning_rate = 0.01, verbose = True):
    for e in range(epochs):
        error = 0 
        for x, y in zip(x_train, y_train):
            #forward
            output = predict(network, x)

            #error
            error += loss(y, output)

            #backward
            gradient = loss_gradient(y, output)
            for layer in reversed(network):
                gradient = layer.backward(gradient, learning_rate)

        error/=len(x_train)
        if verbose:
            print(f'{e+1}/{epochs},error = {error}')
                           
```

```python
def mse(y_true, y_pred):
    return np.mean(np.power(y_true - y_pred, 2))
def mse_gradient(y_true, y_pred):
    return 2 * (y_pred - y_true)/np.size(y_true)
```

```python
AND_network =[
    Dense(2,1)
]
```

```python
x = np.reshape(x,(4,2,1))
```

```python
y=np.reshape(y,(4,1,1))
print(y)
```

```python
train(AND_network,mse,mse_gradient,x, y,epochs = 10000, learning_rate = 0.01, verbose = True)
```

```python
print("AND GATE")
print("0 AND 0 ->",predict(AND_network,[[0],[0]]))
print("0 AND 1 ->", predict(AND_network,[[0],[1]]))
print("1 AND 0 ->", predict(AND_network,[[1],[0]]))
print("1 AND 1 ->", predict(AND_network,[[1],[1]]))
```

```python
data = {
    "a": [0,0,1,1],
    "b": [0,1,0,1],
    "a+b":[0,1,1,1]}
df = pd.DataFrame(data)
```

```python
y
```

```python
OR_network=[
    Dense(2,1)
]
```

```python
x = np.reshape(x,(4,2,1))
y = np.reshape(y,(4,1,1))
```

```python
print("OR GATE")
print("0 OR 0 ->",predict(OR_network,[[0],[0]]))
print("0 OR 1 ->", predict(OR_network,[[0],[1]]))
print("1 OR 0 ->", predict(OR_network,[[1],[0]]))
print("1 OR 1 ->", predict(OR_network,[[1],[1]]))
```

```python
# XOR network -> MULTIPLE LAYER PERCEPTRON
```

```python
data = {
    "a": [0,0,1,1],
    "b": [0,1,0,1],
    "a(+)b":[0,1,1,0]}
df = pd.DataFrame(data)
```

```python
df
```

```python
x = np.array(df.drop('a(+)b',axis=1))
y = np.array(df['a(+)b'])
```

```python
XOR_network = [
    Dense(2,5),
    Dense(5,3),
    Dense(3,1)
]
```

```python
x = np.reshape(x,(4,2,1))
y = np.reshape(y,(4,1,1))
```

```python
train(XOR_network, mse, mse_gradient, x, y, epochs = 10000, learning_rate=0.01, verbose=True)
```

```python
print("XOR GATE")
print("0 XOR 0 ->",predict(XOR_network,[[0],[0]]))
print("0 XOR 1 ->", predict(XOR_network,[[0],[1]]))
print("1 XOR 0 ->", predict(XOR_network,[[1],[0]]))
print("1 XOR 1 ->", predict(XOR_network,[[1],[1]]))
```
